```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
current_directory = os.getcwd() + '/Desktop/Work/iRobot/'
data_file_name = 'customer_data.csv'

df = pd.read_csv(current_directory + data_file_name)
df = df.rename(columns={"Age":"age","Genre":"gender", "Annual Income (k$)": "income", "Spending Score (1-100)":"spending_score"})
df
```

```python
df = df.replace({'gender': {'Male':0, 'Female': 1}})
df = df[['gender', 'age', 'income', 'spending_score']]
df
```

```python
from sklearn.svm import OneClassSVM
nu = 0.05
ocsvm = OneClassSVM(kernel='rbf', gamma='scale', nu=nu)
ocsvm.fit(df)
pred_ocsvm = ocsvm.predict(df)
pred_ocsvm
```

```python
outliers = df[pred_ocsvm == -1]
outliers
```
